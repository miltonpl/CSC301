### Calculate d(cos(x))dx using 4 differt methods, x= 1 radian

```python
import sympy as sy
import numpy as np
from sympy.functions import sin,cos,exp,log
from IPython.display import Image
import matplotlib.pyplot as plt
import math
from math import*

# plt.style.use("ggplot")

#Factorial function
# d(cos(x))/dx  = -sin(x)
def cos_f(x):
    return cos(x)


```

## Three- endpoint formula 


```python
def three_endpoint_formulas(f,h,x0):
    
    result = ((-3*f(x0) + 4*f(x0+h) - f(x0 + 2*h))/(2*h))
    return float("%.16f"%(result))


    
```

## Three- point formula 

```python
def three_point_formulas(f,h,x0):
    
    result = ((f(x0-h) - f(x0-h))/(2*h))
    
    return float("%.16f"%(result))

```

```python
Image("img/5midpoint.png")
```

```python
def five_midpoint_formula(f,h,x0):
    
    result = (f(x0-2*h) - 8*f(x0 - h) + 8*f(x0 + h) - f(x0+ 2*h))/(12*h)
    return float("%.16f"%(result))

```

```python
Image("img/sin.png")
```

```python
def Taylor_Approximation(x,n,result):
    
    for i in range(0, n):
        temp = (((-1)**i)*(x**(2*i+1)))/(math.factorial(2*i+1))
        result += temp
    return -result
        
    
    
```

```python
Image("img/error.png")
    
```

```python
def data_error(actual, measured, msg):
    y= []
    actual = float("%.16f"%(actual))
    measured = float("%.16f"%(measured))
    RELATIVE_ERROR=abs(measured - actual)/abs(actual)
    y.append(RELATIVE_ERROR)
    ABSOLUTE_ERROR =abs(measured - actual)
    
    y.append(ABSOLUTE_ERROR)
   
    print("........................."+msg+"....................")
    print("Actual Result: ",actual)
    print("Measured Result:", measured)
    print("___________________________________________________")
    print("Absolute Error              |       Relative Error")
    print(ABSOLUTE_ERROR,"      |      ",RELATIVE_ERROR)
    
    return y 
    
    
```

## Calculations Below

```python
# f = 1.0 radians
x = 1.0 
h = 0.000000001
ACTUAL_VALUE= - sin(1)

#cos'(x)
three_endpoint_result=three_endpoint_formulas(cos_f,h,x)

msg1="TREE ENDPOINT FORMULA"
error3Endpoint = data_error(ACTUAL_VALUE, three_point_result,msg1)
```

```python
three_point_result=three_endpoint_formulas(cos_f,h,x)

msg2="TREE POINT FORMULA"
error3point = data_error(ACTUAL_VALUE, three_point_result,msg2)
```

```python
five_midpoint_result=five_midpoint_formula(cos_f,h,x)

msg3="FIVE MIDPOINT FORMULA"
error5Midpoint = data_error(ACTUAL_VALUE, five_midpoint_result,msg3)
```

```python
n= 50
taylor_sin=Taylor_Approximation(x,n,0)
msg4="TAYLOR APROXIMATION"
taylor_error = data_error(ACTUAL_VALUE, taylor_sin,msg3)
```

```python
#relaive, abs
print(taylor_error)
print(error5Midpoint)
print(error3point)
print(error3Endpoint)
xlist=["Taylor","FiveMidPoint","TreeMidPoint","ThreeMidPoint"]
y1_relativeError=[taylor_error[0],error5Midpoint[0] ,error3Endpoint[0] ,error3point[0]]
y1_absError=[taylor_error[1],error5Midpoint[1] ,error3Endpoint[1] ,error3point[1]]

```

```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd

s = pd.Series(y1_relativeError, index=xlist)

s.plot.bar(figsize=(10,5))
plt.grid()
plt.xlabel('Method')
plt.ylabel('Error')
plt.title("RELATIVE ERROR");
```

```python

c = pd.Series(y1_absError, index=xlist)

c.plot.bar(figsize=(10,5))
plt.grid()
plt.xlabel('Method')
plt.ylabel('Error')
plt.title("ABSOLUTE ERROR");
```
