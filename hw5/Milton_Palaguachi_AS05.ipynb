```python
#Using the notation h = x1  x0 gives the following rule:
#Trapezoidal Rule:

def Trapezoidal_Rule(x0,x1,f):
    h = x1 -x0
    return (h/2)*(f(x0)+ f(x1))
```

```python
def Simpsons_Rule(f,x0,x2):
    a=x0
    b=x2
    h=(b-a)/2
    x1=a+h
    return(h/3)*(f(x0)+4*f(x1)+f(x2))
    
```

```python
def Open_Newton_Cotes_Formula_n0(f,x0,x1,n):
    #i={0,1,...n}
    a= x0
    b=x1
    h= (b-a)/(n+2)
    x0 = x0+ 0*h
    return(2*h*f(x0,))
    
    
```

```python
def Open_Newton_Cotes_Formula_n1(f,x0,x1,n):
    #i={0,1,...n}
    a = x0
    b = x1
    h=(b-a)/(n+2)
    x0 = x0+ 0*h
    x1 = x0+ 1*h
    return(3*h*(f(x0)+f(x1)))/2
```

```python
Image("img/n2_open_Newton.png")
# the Immages was taken from the book Richard L. Burde, J. Douglas Faires. Chapter 4.
```

```python
def Open_Newton_Cotes_Formula_n2(f,x0,x1,n):
    a = x0
    b = x1
    h=(b-a)/(n+2)
    x0 = x0+ 0*h
    x1 = x0+ 1*h
    x2 = x0+ 2*h

    return(4*h*(2*f(x0)-f(x1)+ 2*f(x2)))/3
```

```python
Image("img/n3_open_Newton.png")
# the Immages was taken from the book Richard L. Burde, J. Douglas Faires. Chapter 4.
```

```python
def Open_Newton_Cotes_Formula_n3(f,x0,x1,n):
    a = x0
    b = x1
    h=(b-a)/(n+2)
    x0 = x0 + 0*h
    x1 = x0 + 1*h
    x2 = x0 + 2*h
    x3 = x0 + 3*h

    return(5*h*(11*f(x0) + f(x1) + f(x2) + 11*f(x3)))/24
```

```python
def Simpsons_approximation(f,x0,x2):
    a=x0
    b=x2
    h=(b-a)/2
    return(h/3)*(f(a)+4*f(a+h)+f(b))
```

```python
Image("img/composite_Simpson.png")
# the Immages was taken from the book Richard L. Burde, J. Douglas Faires. Chapter 4.
```

```python
def Composite_Simpson(f,x0,x2):
    a=x0
    b=x2
    h=(b-a)/2
    return(h/6)*(f(a) + 4*f(a + h/2) + 2*f(a+h) + 4*f(a+3*h/2)+ f(b))
```

```python
def Composite_Trapezoidal_Rule(f,x0,x1,n):
    a=x0
    b=x1
    h=(b-a)/n
    sum=0.0
    for j in range(0,n):
        sum = sum + f(a + j*h)
        
    return(h/2)*(f(a) + 2*sum + f(b))
```

```python
def data_error(msg, measured, actual):
    y= []
    actual = float("%.16f"%(actual))
    measured = float("%.16f"%(measured))
    RELATIVE_ERROR=abs(measured - actual)/abs(actual)*100
    y.append(RELATIVE_ERROR)
    ABSOLUTE_ERROR =abs(measured - actual)*100
    
    y.append(ABSOLUTE_ERROR)
   
    print("........................."+msg+"....................")
    print("Measured Result:", measured)
    print("___________________________________________________")
    print("Absolute Error   %        |       Relative Error %")
    print(ABSOLUTE_ERROR,"       |      ",RELATIVE_ERROR)
    
    return y 
```
